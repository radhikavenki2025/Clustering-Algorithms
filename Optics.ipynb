```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:,[3,4]].values
```

```python
from sklearn.cluster import OPTICS

clustering = OPTICS(min_samples=10).fit(X)

clustering.labels_
```

```python
supervised=pd.DataFrame(dataset)

supervised['Cluster_group']=clustering.labels_

supervised
```

```python
import seaborn as sns

facet=sns.lmplot(data=supervised , x=supervised.columns[3] ,y=supervised.columns[4] , hue=supervised.columns[5],
                 fit_reg=False , legend=True , legend_out=True)
```

```python
#### min_samples less means number of clusters more 

#### min_samples = 2 , number of clusters = 57

#### min_samples = 5 , number of clusters = 10

#### min_samples = 10 , number of clusters = 7 
```
