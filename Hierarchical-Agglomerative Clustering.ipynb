```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
```

```python
dataset
```

```python
X = dataset.iloc[:,[3,4]].values
```

```python
import scipy.cluster.hierarchy as sch

dendrogram = sch.dendrogram(sch.linkage(X, method='ward'))

plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean distances')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

clusmodel= AgglomerativeClustering(n_clusters=5)

label=clusmodel.fit_predict(X)
```

```python
supervised=pd.DataFrame(dataset)
```

```python
supervised['Cluster_group']=label
```

```python
supervised
```

```python
import seaborn as sns

facet=sns.lmplot(data=supervised , x=supervised.columns[3] ,y=supervised.columns[4] , hue=supervised.columns[5],
                 fit_reg=False , legend=True , legend_out=True)
```

```python
#### Customers are segmented based on their Income range and spending pattern

#### Cluster group 0 : Income range between 70K - 140K spend score 0-40 savings is very high

#### Cluster group 1 : Income range between 40K - 80K spend score 30-60 savings is less

#### Cluster group 2 : Income range between 70K - 140K spend score 60-100 savings is very less

#### Cluster group 3 : Income range between 10K - 40K spend score 60 - 100 savings is NIL

#### Cluster group 4 : Income range between 10K - 40K spend score 0 - 40 savings is NIL
```
